```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3,4,5,6,7])
y=np.array([78.4,78.6,78.5,78.9,78.2,79.4,78.6])
plt.plot(x, y, color="r",marker="o")
plt.grid()
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(-5,5,300)
y1 = x **2
y2 = (x-2) **2
plt.plot(x, y1, color="r")
plt.plot(x, y2, color="k", linestyle ="--")
plt.grid()
plt.show
```

```python
x=np.linspace(-5,5)
y1 = 2*x
y2 = 5*x

plt.plot(x, y1, label="y1 value", color="r")
plt.plot(x, y1, label="y2 value",linestyle="dashed", color="k")

plt.xlabel("x-axis", size=14)
plt.ylabel("y-axis", size=14)
plt.title("name of graph")
plt.grid()
plt.legend()

plt.show()
```
